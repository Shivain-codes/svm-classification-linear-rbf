```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from matplotlib.colors import ListedColormap

# 1️⃣ Load dataset (Breast Cancer dataset - for simplicity let’s use two features for 2D visualization)
cancer = datasets.load_breast_cancer()
X = cancer.data[:, :2]  # Using first two features for 2D plot
y = cancer.target

# 2️⃣ Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3️⃣ Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4️⃣ Train linear SVM
svm_linear = SVC(kernel='linear', C=1.0, random_state=42)
svm_linear.fit(X_train_scaled, y_train)

# 5️⃣ Train RBF SVM
svm_rbf = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_rbf.fit(X_train_scaled, y_train)

# 6️⃣ Function to plot decision boundary
def plot_decision_boundary(model, X, y, title):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#0000FF'])
    
    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=cmap_light)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# 7️⃣ Visualize decision boundaries
plot_decision_boundary(svm_linear, X_train_scaled, y_train, "Linear SVM Decision Boundary (Train)")
plot_decision_boundary(svm_rbf, X_train_scaled, y_train, "RBF SVM Decision Boundary (Train)")

# 8️⃣ Evaluate models
print("Linear SVM Accuracy:", accuracy_score(y_test, svm_linear.predict(X_test_scaled)))
print("RBF SVM Accuracy:", accuracy_score(y_test, svm_rbf.predict(X_test_scaled)))
print("\nClassification Report (RBF SVM):\n", classification_report(y_test, svm_rbf.predict(X_test_scaled)))

# 9️⃣ Hyperparameter tuning (C and gamma for RBF SVM)
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.01, 0.1, 1, 10]
}

grid = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=5, verbose=1, n_jobs=-1)
grid.fit(X_train_scaled, y_train)

print("Best Params from GridSearch:", grid.best_params_)
print("Best CV Score:", grid.best_score_)

#  🔟 Visualize best model
best_rbf = grid.best_estimator_
plot_decision_boundary(best_rbf, X_train_scaled, y_train, "Best RBF SVM Decision Boundary (Train)")
```
